```python
# word cloud
import numpy as np
import pandas as pd
from nltk.corpus import stopwords
```

```python
word_string = ' '.join(inaugural.words(fileids='2017-Trump.txt'))
```

```python
import wordcloud
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
wordcloud = WordCloud(stopwords=STOPWORDS, background_color='white', width=1250,height=1050).generate(word_string)
```

```python
wordcloud
```

```python
plt.figure(figsize=(15,10))
plt.clf()
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
